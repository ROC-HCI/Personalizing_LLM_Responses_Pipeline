```python
import os
import zipfile
import json
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Path to the folder containing the .zip files
folder_path = 'path'  # Change this to the correct path

# List to store data from json files
data_list = []

df = pd.DataFrame()

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.zip'):
        zip_path = os.path.join(folder_path, filename)

        # Unzipping the .zip file
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(folder_path)

        # Loop through extracted files
        for extracted_file in zip_ref.namelist():
            if extracted_file.endswith('.json'):
                json_file_path = os.path.join(folder_path, extracted_file)

                # Read JSON file and append to data_list
                with open(json_file_path, 'r') as json_file:
                    try:
                        json_data = json.load(json_file)
                        data_list.append(json_data)  # Assuming each json contains a dictionary
                        df_dictionary = pd.DataFrame.from_dict(json_data)
                        # print(df_dictionary)
                        df = pd.concat([df, df_dictionary], ignore_index=True)
                    except json.JSONDecodeError as e:
                        print(f"Error reading {json_file_path}: {e}")


# # Save the DataFrame to a CSV file
df.to_csv('articles_compiled.csv', index=False)

# print(f"Data has been saved to {output_csv_path} and .json files have been deleted.")
```
